# Pre-Processing

### 1. Converting to lowercase

```python
df["review"] = df["review"].str.lower()
```

### 2. Removing the URls

```python
import re
```

```python
def remove_urls(text):
    re.sub(r"http\S+", "", text) # ( pattern, replacement, string)


df["review"] = df["review"].apply(remove_urls)
```
